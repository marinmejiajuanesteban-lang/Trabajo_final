```python
# El dataset no incluye nombres de columnas en el archivo original,
# por lo que se asignan manualmente segun la documentación oficial

columnas = [
    "edad", "tipo_trabajo", "peso_muestral", "educacion", "educacion_num",
    "estado_civil", "ocupacion", "relacion", "raza", "sexo",
    "ganancia_capital", "perdida_capital", "horas_semana", "pais_origen", "ingreso"
]

df.columns = columnas

df.head()
```

```python
for col in df.select_dtypes(include="object"):
    print(f"\nColumna: {col}")
    print(df[col].unique())
```

Se revisaron las variables categóricas y se encontraron valores ‘?’ que representan datos faltantes. También se identificaron espacios al inicio de algunos valores, lo que puede causar errores en el análisis

```python
#Eliminar espacios al inicio de los valores
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip()

# Reemplazar '?' por valores nulos
df.replace("?", np.nan, inplace=True)
df.isnull().sum()

#Verificar
(df == "?").sum()
df["ingreso"].unique()

```
